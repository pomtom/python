```python
import pandas as pd
import pyodbc

# Import CSV
data = pd.read_csv (r'D:\temp\banknote-authentication.csv')   
df = pd.DataFrame(data)

# Connect to SQL Server
conn = pyodbc.connect('Driver=SQL Server;'
                      'Server=BTL096;'
                      'Database=BANK;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()

# Create Table
cursor.execute('''
CREATE TABLE [dbo].[banknote](
   [variance] [real] NULL,
   [skew] [real] NULL,
   [curtosis] [real] NULL,
   [entropy] [real] NULL,
   [classa] [real] NULL
) ON [PRIMARY]
               ''')

# Insert DataFrame to Table
for row in df.itertuples():
    cursor.execute('''
                INSERT INTO banknote (variance,skew,curtosis,entropy,classa)
                VALUES (?,?,?,?,?)
                ''',
                row.variance, 
                row.skew,
                row.curtosis,
                row.entropy,
                row.classa
                )
conn.commit()
```
